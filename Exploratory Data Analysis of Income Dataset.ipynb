## 1. Data Preparation

### 1.1 Observing Data and Labelling the Attributes

```python
#Importing the dataset and the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('adult.data', header = None)
```

```python
df.head(5)
```

```python
#Adding headers to the dataset
df.columns = ['Age','Workclass','Fnlwgt','Education','Education-num','Marital-Status','Occupation','Relationship','Race','Sex','Capital-gain','Capital-loss','hours-per-week','Native-Country','Income']
```

```python

df.head(10)
```

```python
#Viewing the shape of the data 
df.shape
```

```python
#Replacing all ? values with Nan
import numpy as np
df.replace(" ?", np.nan, inplace = True)
df.head(15)
```

```python
# Displaying number of null values for each of the columns. True indicates
# null values 

missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

## 2. Correlation between numeric variables

```python
# Displaying a heatmap to show correlation between the numeric variables
corr = df.corr()
fig6 = sns.heatmap(corr,annot = True ,xticklabels=corr.columns,yticklabels=corr.columns,cmap="RdPu").set_title('Correlation between numeric variables')
```
